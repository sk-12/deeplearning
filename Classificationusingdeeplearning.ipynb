```
import pandas as pd
import numpy as np
```

```
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```
data.isnull().sum()
```

```
x=data.iloc[:,3:-1]
y=data.iloc[:,-1]
```

```
geography=pd.get_dummies(data['Geography'],drop_first=True)
gender=pd.get_dummies(data['Gender'],drop_first=True)
```

```
x=x.drop(['Geography','Gender'],axis=1)
```

```
x=pd.concat([x,geography,gender],axis=1)
```

```
x.head()
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```
from sklearn.preprocessing import StandardScaler
object1=StandardScaler()
x_train=object1.fit_transform(x_train)
x_test=object1.transform(x_test)
```

```
import tensorflow as tf
```

```
from tensorflow.keras import Sequential
```

```
object2=Sequential()
object2.add(tf.keras.layers.Dense(units=10,input_dim=11,kernel_initializer='he_normal',activation='relu'))
object2.add(tf.keras.layers.Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
object2.add(tf.keras.layers.Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
object2.add(tf.keras.layers.Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```
object2.summary()
```

```
object2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
object2.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```
y_pred=object2.predict(x_test)
y_pred=(y_pred>0.5)
from sklearn.metrics import confusion_matrix,accuracy_score
object3=accuracy_score(y_pred,y_test)
object4=confusion_matrix(y_pred,y_test)
```
